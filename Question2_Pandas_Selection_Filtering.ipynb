Creating the Pandas DataFrame

```python
import pandas as pd

data = {
    "EmployeeID": [201,202,203,204,205,206,207,208,209,210],
    "Name": ["Sarah", "Michael", "Jennifer", "David", "Lisa", "James", "Emma", "Robert", "Sophia", "William"],
    "Department": ["Engineering", "Sales", "Marketing", "Engineering", "HR", "Sales", "Engineering", "Marketing", "Sales", "HR"],
    "Salary": [85000,62000,71000,92000,58000,68000,78000,75000,65000,60000],
    "Performance Score": [4,3,5,5,4,4,3,4,5,3],
    "Years of Service": [5,3,7,8,4,6,2,5,4,3]
}

df = pd.DataFrame(data)
df.info()
```

Displaying only 'Name', 'Department', and 'Salary' columns

```python
df[["Name", "Department", "Salary"]]
```

displaying the employee with EmployeeID 208 using .loc()

```python
df.loc[df["EmployeeID"]==208]
```

```python
df.loc[7]
```

Filter and displaying all employees from the 'Engineering' department

```python
df[df["Department"] == "Engineering"]
```

Filter and displaying all employees with Salary greater than 70000

```python
df[df["Salary"] > 70000]
```

Filter and displaying employees from 'Sales' department with Performance Score >= 4

```python
df[(df["Department"] == "Sales") & (df["Performance Score"] >= 4)]
```

Sorting the entire DataFrame by Salary in descending order and displaying the top 3 highest-paid employees

```python
result = df.sort_values("Salary", ascending=False)
result.head(3)
```

Sorting by Department (ascending) and then by Salary (descending) within each department

```python
df.sort_values(["Department","Salary"], ascending=[True,False])
```
