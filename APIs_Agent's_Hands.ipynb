# Task 1

```python
import requests
import pandas as pd

url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
  headers = {"User-Agent": "Mozilla/5.0"}
  response = requests.get(url, headers=headers)
  print(f"Status Code: {response.status_code}")

  if response.status_code == 200:
    data = response.json()
    print(f"Total countries retrived: {len(data)}")
  else:
    print(f"Error: Failed to fetch data. Status Code {response.status_code}")
    data = []

except requests.exceptions.RequestException as e:
  print(f"Request failed: {e}")
  data = []
```

# Task 2

```python
records = []

for country in data:
  record = {
      "common_name": country.get("name", {}).get("common", "N/A"),
      "population": country.get("population", 0),
      "region": country.get("region", "N/A"),
      "area": country.get("area", 0)
  }
  records.append(record)

df = pd.DataFrame(records)

print("\nFirst 10 rows of the DataFrame: ")
df.head(10)
```

# Task 3

```python

region_population = df.groupby("region")["population"].sum()
highest_population_region = region_population.idxmax()

largest_country = df.loc[df["area"].idxmax(), "common_name"]

print("\nSummary: ")
print(f"Region with highest total population: {highest_population_region}")
print(f"Country with the largest area: {largest_country}")
```
