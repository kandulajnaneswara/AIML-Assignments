# Task 1: Create Database


```python
from IPython.utils.path import random
from sqlite3.dbapi2 import connect
import pandas as pd
import sqlite3

conn = sqlite3.connect("employees.db")
cursor = conn.cursor()

# Creating the Employees database
cursor.execute("""CREATE TABLE IF NOT EXISTS employees(
  emp_id INTEGER PRIMARY KEY AUTOINCREMENT,
  name   TEXT NOT NULL,
  department  TEXT NOT NULL,
  salary   REAL NOT NULL,
  years_experience INTEGER,
  performance_score REAL
)
""")

departments = ["Engineering", "Sales", "Marketing", "HR", "Finance"]

# Generating employee data
emp_data = []

for i in range(1,41):
  name = f"Employee_{i}"
  department = random.choice(departments)
  salary = random.randint(50000, 150000)
  years_experience = random.randint(1, 15)
  performance_score = round(random.uniform(1.0, 5.0), 2)

  emp_data.append((i, name, department, salary, years_experience, performance_score))

cursor.executemany("""INSERT INTO employees
(emp_id, name, department, salary, years_experience, performance_score) VALUES (?, ?, ?, ?, ?, ?)
""", emp_data)

conn.commit()

cursor.execute("SELECT * FROM employees")
rows = cursor.fetchall()

print("Employee records: ")
for row in rows:
  print(row)

conn.close()

```

# Task 2: SQL Queries

Query 1:

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("employees.db")

query1 = """SELECT name, department, salary, performance_score
FROM employees
WHERE performance_score >= 4.0 AND years_experience >= 3
ORDER BY performance_score DESC
LIMIT 15"""
df_query1 = pd.read_sql(query1, conn)

print("\nQuery1 Results: ")
print(df_query1)
```

Query 2:

```python
query2 = """
SELECT *
FROM employees
WHERE salary BETWEEN 70000 AND 110000
AND department IN ('Engineering', 'Sales')
ORDER BY department, salary DESC
"""
df_query2 = pd.read_sql(query2, conn)

print("\nQuery2 Results: ")
print(df_query2)
```

Query 3:

```python
query3 = """
SELECT department,
COUNT(*) AS employee_count,
AVG(salary) AS avg_salary
FROM employees
GROUP BY department
ORDER BY avg_salary DESC
"""

df_query3 = pd.read_sql(query3, conn)

print("\nQuery 3 Results:")
print(df_query3)

conn.close()
```

# Task 3: Pandas Analysis

question 1:

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("employees.db")

employees_df = pd.read_sql("SELECT * FROM employees", conn)

avg_performance_score = employees_df.groupby("department")["performance_score"].mean()

print("\nAverage Performance Score by Department: ")
print(avg_performance_score)
```

```python
query1_pandas = employees_df.loc[(employees_df["performance_score"] >= 4.0) & (employees_df["years_experience"] >= 3),["name", "department", "salary", "performance_score"]].sort_values(by="performance_score", ascending=False).head(15)

print("\nQuery 1 using Pandas:")
print(query1_pandas)
```

question 3:

SQL and Pandas are both powerful tools for data analysis, but they use different syntaxes and are suited to different environments.

SQL is a declarative query language used to retrieve and manipulate data stored in relational databases. Its syntax includes operators like =, logical conditions such as AND, and sorting commands like ORDER BY.

Pandas is a Python library designed for in-memory data analysis. It uses Python-style operators such as == for comparison, & for combining conditions, and functions like .sort_values() for sorting.

SQL is typically used when working directly with databases, especially when the data resides on a server or must be processed close to the storage layer.

Pandas is more flexible for data exploration, transformation, and machine learning workflows because it integrates with the Python ecosystem.

For datasets that are too large to fit into memory, SQL-based systems are generally better because the database engine processes data on disk and optimizes queries using indexes and query planners. Pandas, on the other hand, loads data into memory, which can limit scalability for extremely large datasets. Therefore, SQL is preferred for large-scale data storage and retrieval, while Pandas excels in detailed analysis and modeling.
